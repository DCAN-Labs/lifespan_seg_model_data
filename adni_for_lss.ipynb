```python
import pandas as pd
```

```python
# load adni-freesurfer_3_11_2025.csv and final_adni_qc.csv as dataframes
freesurfer = pd.read_csv('adni-freesurfer_3_11_2025.csv')
qc = pd.read_csv('UCSFFSL_02_01_16_27Mar2025.csv')
#qc = pd.read_csv('UCSFFSX_11_02_15_27Mar2025.csv')
freesurfer
```

```python
qc
```

```python
# drop rows in qc df where 'OVERALLQC' is not 'Pass'
qc = qc[qc['OVERALLQC'] == 'Pass']
qc
```

```python
# drop rows in qc df where there are NaN values in 'RID' or 'VISCODE'
qc = qc.dropna(subset=['RID', 'VISCODE'])
qc
```

```python
# in subject column for freesurfer, keep only characters that come after second underscore, and convert to int64
freesurfer['Subject'] = freesurfer['Subject'].str.split('_').str[2].astype('int64')
freesurfer
```

```python
freesurfer.dtypes
```

```python
# rename Subject column to RID and Visit column to VISCODE2
freesurfer = freesurfer.rename(columns={'Subject':'RID', 'Visit':'VISCODE'})
freesurfer
```

```python
# print dtypes for both dataframes
print(freesurfer.dtypes)
print(qc.dtypes)
```

```python
# merge freesurfer and qc dataframes on RID and VISCODE2
merged = pd.merge(qc, freesurfer, on=['RID', 'VISCODE'])
merged
```
