```python
import pandas as pd
```

```python
# load adni-freesurfer_3_11_2025.csv and final_adni_qc.csv as dataframes
freesurfer = pd.read_csv('../data-for-lssm/raw/adni-freesurfer_3_11_2025.csv')
qc1 = pd.read_csv('../data-for-lssm/raw/UCSFFSL_02_01_16_27Mar2025.csv')
qc2 = pd.read_csv('../data-for-lssm/raw/UCSFFSX_11_02_15_27Mar2025.csv')
freesurfer
```

```python
qc1
```

```python
qc2
```

```python
# drop rows in qc df where 'OVERALLQC' is not 'Pass'
qc1 = qc1[qc1['OVERALLQC'] == 'Pass']
qc2 = qc2[qc2['OVERALLQC'] == 'Pass']
qc1
```

```python
qc2
```

```python
# drop rows in qc df where there are NaN values in 'RID' or 'VISCODE'
qc1 = qc1.dropna(subset=['RID', 'VISCODE'])
qc2 = qc2.dropna(subset=['RID', 'VISCODE'])
qc1
```

```python
qc2
```

```python
# keep only the following columns in qc df: 'RID', 'VISCODE', 'OVERALLQC'
qc1 = qc1[['RID', 'VISCODE', 'OVERALLQC']]
qc2 = qc2[['RID', 'VISCODE', 'OVERALLQC']]
qc1
```

```python
# merge qc1 and qc2 on 'RID' and 'VISCODE'
qc = pd.merge(qc1, qc2, on=['RID', 'VISCODE'], how='outer')
qc
```

```python
# if qc[OVERALLQC_x] == 'Pass' or qc[OVERALLQC_y] == 'Pass' then qc[OVERALLQC] = 'Pass'
qc['OVERALLQC'] = qc.apply(lambda x: 'Pass' if x['OVERALLQC_x'] == 'Pass' or x['OVERALLQC_y'] == 'Pass' else 'Fail', axis=1)
qc
```

```python
# drop columns 'OVERALLQC_x' and 'OVERALLQC_y'
qc = qc.drop(columns=['OVERALLQC_x', 'OVERALLQC_y'])
qc
```

```python
# in subject column for freesurfer, keep only characters that come after second underscore, and convert to int64
freesurfer['Subject'] = freesurfer['Subject'].str.split('_').str[2].astype('int64')
freesurfer
```

```python
freesurfer.dtypes
```

```python
# rename Subject column to RID and Visit column to VISCODE2
freesurfer = freesurfer.rename(columns={'Subject':'RID', 'Visit':'VISCODE'})
freesurfer
```

```python
# print dtypes for both dataframes
print(freesurfer.dtypes)
print(qc.dtypes)
```

```python
# merge freesurfer and qc dataframes on RID and VISCODE2
merged = pd.merge(qc, freesurfer, on=['RID', 'VISCODE'])
merged
```

```python
# keep only the following columns in merged df: 'RID', 'VISCODE', 'OVERALLQC', 'Age', 'Description'
merged = merged[['RID', 'VISCODE', 'OVERALLQC', 'Age', 'Description']]
# convert 'Description' column to string
merged['Description'] = merged['Description'].astype(str)
merged
```

```python
# keep only rows that have description that has value of 'FreeSurfer Cross-Sectional Processing aparc+aseg'
merged = merged[merged['Description'] == 'FreeSurfer Cross-Sectional Processing aparc+aseg']
merged
```

```python
# drop duplicates in merged df
merged = merged.drop_duplicates()
merged
```

```python
# keep only the following columns in merged df: 'RID', 'VISCODE', 'Age'
merged = merged[['RID', 'VISCODE', 'Age']]
merged
```

```python
# drop duplicated rows in merged df
merged = merged.drop_duplicates(subset=['RID', 'VISCODE'])
merged = merged.dropna(subset=['Age'])
merged
```

```python
# value counts for age
age_counts = merged['Age'].value_counts()
# sort the value counts
age_counts = age_counts.sort_index()
age_counts
```

```python
# save to csv file for df with age column and one without
merged.to_csv('../data-for-lssm/cleaned/adni_final_qc_with_age.csv', index=False)
merged_no_age = merged.drop(columns=['Age'])
merged_no_age.to_csv('../data-for-lssm/cleaned/adni_final_qc_no_age.csv', index=False)
```
