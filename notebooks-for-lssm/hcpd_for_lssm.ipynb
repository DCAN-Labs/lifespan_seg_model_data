```python
import pandas as pd
```

```python
hcpd_data = pd.read_csv('../data-for-lssm/bl-hcp-d_s3_status_report.csv')
hcpd_data
```

```python
hcpd_data_cleaned = hcpd_data[['subj_id','ses_id','structural']]
hcpd_data_cleaned.dtypes
```

```python
# only keep rows with structural data == 'ok'
hcpd_data_cleaned = hcpd_data_cleaned[hcpd_data_cleaned['structural'] == 'ok']
hcpd_data_cleaned = hcpd_data_cleaned[['subj_id','ses_id']]
hcpd_data_cleaned['subj_id'] = hcpd_data_cleaned['subj_id'].astype(str)
hcpd_data_cleaned['ses_id'] = hcpd_data_cleaned['ses_id'].astype(str)
```

```python
# check for duplicates
hcpd_data_cleaned = hcpd_data_cleaned.drop_duplicates()
hcpd_data_cleaned = hcpd_data_cleaned.reset_index(drop=True)
hcpd_data_cleaned
```

```python
# save cleaned data
hcpd_data_cleaned.to_csv('../data-for-lssm/hcpd_data_no_age.csv', index=False)
```
