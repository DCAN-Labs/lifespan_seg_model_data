```python
import pandas as pd
```

```python
ideas_df = pd.read_csv('../data-for-lssm/raw/ideas_Metadata_Release_Anon.csv')
ideas_df
```

```python
ideas_con_df = pd.read_csv('../data-for-lssm/raw/ideas_Metadata_Controls_Release.csv')
ideas_con_df
```

```python
# keep only the columns ID, Op_Side, OP_Type, and Binned_Age_at_Scan from ideas_df
ideas_df = ideas_df[['ID', 'Op_Side', 'Op_Type', 'Binned_Age_at_Scan']]
# keep only the columns ID and Binned_Age_at_Scan from ideas_con_df
ideas_con_df = ideas_con_df[['ID', 'Binned_Age_at_Scan']]
```

```python
ideas_df
```

```python
ideas_df['Op_Type'].value_counts()
```

```python
# for values in the Binned_Age_at_Scan column in ideas_con_df that are not in the format 'x to y', set the value to y
def convert_binned_age(value):
    if ' to ' in value:
        return value
    else:
        return value.split(' ')[-1]
    
ideas_con_df['Binned_Age_at_Scan'] = ideas_con_df['Binned_Age_at_Scan'].apply(convert_binned_age)
ideas_con_df.value_counts()
```

```python
# Function to calculate average age from binned age string
def calculate_age_average(age_range):
    if ' to ' in age_range:
        start, end = age_range.split(' to ')
        return (int(start) + int(end)) / 2
    else:
        # If there's no range, just return the single value
        return int(age_range)

ideas_con_df['Binned_Age_at_Scan_avg'] = ideas_con_df['Binned_Age_at_Scan'].apply(calculate_age_average)
ideas_df['Binned_Age_at_Scan_avg'] = ideas_df['Binned_Age_at_Scan'].apply(calculate_age_average)

ideas_con_df['Binned_Age_at_Scan_avg'] = ideas_con_df['Binned_Age_at_Scan_avg'].astype(int)
ideas_df['Binned_Age_at_Scan_avg'] = ideas_df['Binned_Age_at_Scan_avg'].astype(int)
ideas_con_df
```

```python
# provide value counts for the Binned_Age_at_Scan_avg column in ideas_df and ideas_con_df
ideas_df['Binned_Age_at_Scan_avg'].value_counts().sort_index()
```
