```python
import pandas as pd
```

```python
hcpya_1 = pd.read_csv('../data-for-lssm/uncleaned/btc-hcp-ya_s3_status_report.csv')
hcpya_2 = pd.read_csv('../data-for-lssm/uncleaned/HCP_YA_s3_status_report.csv')
hcpya_1
```

```python
hcpya_2
```

```python
hcpya_1_cleaned = hcpya_1[['subj_id','ses_id','structural']]
hcpya_2_cleaned = hcpya_2[['subj_id','ses_id','structural']]
hcpya_1_cleaned
```

```python
hcpya_2_cleaned
```

```python
# merge the two dataframes
hcpya_merged = pd.merge(hcpya_1_cleaned, hcpya_2_cleaned, on=['subj_id', 'ses_id'], how='outer', suffixes=('_1', '_2'))
hcpya_merged
# check for duplicates on subj_id and ses_id
#hcpya_merged['duplicate'] = hcpya_merged.duplicated(subset=['subj_id', 'ses_id'], keep=False)
#hcpya_merged['duplicate'].value_counts()
```

```python
# save out new dataframe for subj_id and ses_id that have "ok" in either structural column
hcpya_merged['structural_1'] = hcpya_merged['structural_1'].str.lower()
hcpya_merged['structural_2'] = hcpya_merged['structural_2'].str.lower()
hcpya_data_cleaned = hcpya_merged[(hcpya_merged['structural_1'] == 'ok') | (hcpya_merged['structural_2'] == 'ok')]
hcpya_data_cleaned
```

```python
# save out only subj_id and ses_id
hcpya_data_cleaned = hcpya_data_cleaned[['subj_id', 'ses_id']]
hcpya_data_cleaned = hcpya_data_cleaned.drop_duplicates()
hcpya_data_cleaned = hcpya_data_cleaned.reset_index(drop=True)
hcpya_data_cleaned
```

```python
# save out the cleaned data
hcpya_data_cleaned.to_csv('../data-for-lssm/cleaned/hcpya__data_no_age.csv', index=False)
# count the number of rows in the cleaned data
hcpya_data_cleaned.shape[0]
```
